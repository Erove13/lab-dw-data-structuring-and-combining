```python
#funcions previous lab
def standarized_columns(df):
    df.columns = df.columns.str.lower().str.replace(" ", "_")
    return df


def cleaning_columns(df):
  #gender column
  df['gender'] = df['gender'].str.lower().map({'f': 'F', 'm': 'M'})
  df['gender'] = df['gender'].fillna('Unknown')


  #state column
  state_map = {'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'}
  df['state'] = df['state'].map(state_map).fillna(df['state'])

  #level education column
  df['education'] = df['education'].str.lower().replace('bachelors', 'bachelor')

  #remove % - customer lifetime value column
  df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '', regex=False)
  df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'])

  #vehicle classes as "Luxury" column
  luxury_classes = ['sports car', 'luxury suv', 'luxury car']
  df['vehicle_class'] = df['vehicle_class'].str.lower().apply(lambda x: 'Luxury' if x in luxury_classes else x)

# number of complaints
  df['number_of_open_complaints'] = df['number_of_open_complaints'].str.split('/').str[1].astype(float)
  return df

def clean_rows(df):
    #Clean those rows
    df.dropna(how="all", inplace =True)
    return df
```

```python
#work with file1
standarized_columns(df_insurance_company1)
```

```python
df_insurance_company1= df_insurance_company1.rename(columns={'st': 'state'})
```

```python
clean_rows(df_insurance_company1)
cleaning_columns(df_insurance_company1)
```

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Your code goes here
url_m = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'
df_marketing = pd.read_csv(url_m)
```

```python
# Group by 'policy_type' and 'month'
df_num_complaints = df_marketing.groupby(['policy_type', 'month']).size().unstack(fill_value=0)

df_num_complaints
```

```python
df_marketing_melted = pd.melt(df_num_complaints.reset_index(), id_vars=['policy_type'], var_name='month', value_name='number_of_complaints')

df_marketing_melted
```
